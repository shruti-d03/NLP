```python
from spacy.tokens import Span  

# Create a span (entity) from token index 0 to 1
# This will take the first token (index 0) from the doc
# label="ORG" → assigning it as an Organization
s1 = Span(doc, 0, 1, label="ORG")  

# Create another span from token index 5 to 6
# This will take the 6th token (index 5)
# label="ORG" → assigning it as an Organization
s2 = Span(doc, 5, 6, label="ORG")  

# Set these spans as named entities in the document
# [s1, s2] → list of custom entities
# default="unmodified" → keeps existing entities unchanged
doc.set_ents([s1, s2], default="unmodified")  
```
