```python
import spacy

# Load the French language model
nlp = spacy.load("fr_core_news_sm")

# Process a French sentence — "Tesla Inc is going to acquire Twitter for $45 billion dollars"
doc = nlp("Tesla Inc va racheter Twitter pour $45 milliards de dollars")

# Loop through named entities found in the sentence
for ent in doc.ents:
    print(ent.text,        # The actual entity text e.g. "Tesla Inc"
          " | ",
          ent.label_,      # Entity category label e.g. "ORG", "MONEY"
          " | ",
          spacy.explain(ent.label_))  # Human-readable explanation of the label

#Tesla Inc  |  ORG  |  Companies, agencies, institutions
#Twitter  |  ORG  |  Companies, agencies, institutions
#$45 milliards de dollars  |  MONEY  |  Monetary values, including unit
```

## Adding a Component to a Blank Pipeline

```python

source_nlp = spacy.load("en_core_web_sm")

nlp = spacy.blank("en")
nlp.add_pipe("ner", source=source_nlp)
nlp.pipe_names
```

```python
doc = nlp("Tesla Inc is going to acquire twitter for $45 billion")
for ent in doc.ents:
    print(ent.text, ent.label_)
```
