```python
from sklearn.feature_extraction.text import CountVectorizer
# Import CountVectorizer from sklearn
# It converts text into numerical vectors using word frequency

v = CountVectorizer()
# Create an object of CountVectorizer
# By default:
# - Converts text to lowercase
# - Ignores punctuation
# - Tokenizes words
# - Builds vocabulary

v.fit(["Thor Hathodawala is looking for a job"])
# 'fit' learns all unique words (vocabulary) from the given sentence
# It assigns an index to each unique word

v.vocabulary_
# Returns a dictionary:
# key = word
# value = index assigned to that word
```

```python
v = CountVectorizer(ngram_range=(1,2))
# Create CountVectorizer object
# ngram_range=(1,2) means:
# 1 в†’ Unigrams (single words)
# 2 в†’ Bigrams (pairs of consecutive words)(2,2)

v.fit(["Thor Hathodawala is looking for a job"])
# Learns vocabulary from the sentence
# Includes both single words and word pairs

v.vocabulary_
# Returns dictionary of words + bigrams with index values
```
