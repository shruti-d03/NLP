```python
# ============================================================
# SPACY - Sentence Tokenization
# ============================================================
import spacy

nlp = spacy.load("en_core_web_sm")   # load the English model

text = "Dr.Strange loves pav bhaji. Hulk loves chaat of Delhi."

doc = nlp(text)          # process the text — creates a Doc object

for sent in doc.sents:   # .sents gives sentence spans
    print(sent)

# Output:
# Dr.Strange loves pav bhaji.
# Hulk loves chaat of Delhi.

# WHY spaCy is smart: it knows "Dr." is an abbreviation,
# NOT a sentence ending. It uses a trained model to decide.
```

```python
# ============================================================
# NLTK - Sentence Tokenization
# ============================================================
import nltk
from nltk.tokenize import sent_tokenize

text = "Dr.Strange loves pav bhaji. Hulk loves chaat of Delhi."

sentences = sent_tokenize(text)   # returns a list of sentence strings
print(sentences)

# Output:
# ['Dr.Strange loves pav bhaji.', 'Hulk loves chaat of Delhi.']

# NLTK also handles "Dr." correctly using the Punkt algorithm
# which learns abbreviation patterns from training data.
```

```python
import nltk
nltk.download('punkt')
nltk.download('punkt_tab')  # needed in newer NLTK versions

from nltk.tokenize import sent_tokenize
sent_tokenize("Dr.Strange loves pav bhaji of mumbai. Hulk loves chaat of Delhi")
```

```python
# ============================================================
# NLTK - Word Tokenization
# ============================================================
import nltk
from nltk.tokenize import word_tokenize

tokens = word_tokenize("Hulk loves chaat of Delhi!")
print(tokens)

# Output: ['Hulk', 'loves', 'chaat', 'of', 'Delhi', '!']
# Returns a plain Python list of strings — simple and easy
```
